# <span style="color:darkorchid">***Exploratory Data Analysis on the datasets***</span>

### <span style="color:darkred">***merging the two data set for the analysis***</span>

```python
import pandas as pd

try:
    # the two csv files 
    artworks_file_path = 'Artworks.csv'
    artists_file_path = 'Artists.csv'

    # Read the CSV files into pandas DataFrames
    artworks_df = pd.read_csv(artworks_file_path)
    artists_df = pd.read_csv(artists_file_path)

    # Convert the key column to the same data type in both DataFrames
    key_column = ['ConstituentID']
    artworks_df[key_column] = artworks_df[key_column].astype(str)
    artists_df[key_column] = artists_df[key_column].astype(str)

    # Merge DataFrames based on the common key ("ConstituentID")
   
    merged_df = pd.merge(artworks_df, artists_df, on='ConstituentID', how='inner')

    # Save the merged DataFrame to a new CSV file
    merged_df.to_csv('merged_file.csv', index=False)

    print("Files merged successfully!")

except Exception as e:
    print(f"An error occurred: {e}")
```

### <span style="color:darkred">***Temporal Trends and Correlation Analysis:***</span>

#### <span style="color:green">***We'll visualize the temporal trends by plotting the number of artworks created each year and explore the classification in which factor(Architecture,Design).***</span>

```python
remove_data['Year'] = pd.to_datetime(remove_data['BeginDate_y']).dt.year

# Get counts of each classification for each year
category_counts_per_year = remove_data.groupby(['Year', 'Classification']).size().reset_index(name='Count')

# Create a bar graph
plt.figure(figsize=(14, 8))
sns.barplot(x='Classification', y='Count', hue='Year', data=category_counts_per_year, palette='viridis')
plt.xticks(rotation=90, ha='right')

# Add labels and title
plt.xlabel('Classification')
plt.ylabel('Number of Artworks')
plt.title('Bar Graph of Classification Counts Over the Years')

# Show the plot
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### The bar graph provides a comprehensive view of the distribution of artworks across different classifications over the years. Notably, 1970 stands out with a significant increase in the number of artworks, particularly in the Photograpy. This suggests a trend or shift in artistic preferences during that period. Additionally, the consistent representation of illustrated book across multiple years indicates its enduring popularity. The visualization offers valuable insights into the evolving landscape of artistic classifications, highlighting distinct patterns and trends over the analyzed years.

For instance, photographshows a significant increase in the number of artists born in the early 20th century, suggesting a concentration of artistic activity during that era. The heatmap effectively visualizes the interplay between artists' birth years and their respective classifications, offering insights into historical trends and artistic movements within the dataset.

### <span style="color:darkred">***Artists' Careers***</span>

### <span style="color:green">***Artists'Analyzing career duration in the years***</span>

```python
remove_data['BeginDate_y'] = pd.to_datetime(remove_data['BeginDate_y'])
remove_data['EndDate_y'] = pd.to_datetime(remove_data['EndDate_y'])

# Calculate the career duration in years
remove_data['career_duration'] = (remove_data['EndDate_y'] - remove_data['BeginDate_y']).dt.days / 365.25

# Create a histogram for career duration
plt.figure(figsize=(10, 6))
ax = sns.histplot(remove_data['career_duration'].dropna(), kde=True, bins=30, color='skyblue')

# Set x-axis labels to display in integer format
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Add labels and title
plt.title('Distribution of Artist Career Durations')
plt.xlabel('Career Duration (years)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
